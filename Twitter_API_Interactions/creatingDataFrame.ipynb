```python
import pandas as pd
```

```python
bh_df = pd.read_csv('misinfoNarratives/ballotHarvesting/locationResults/results.csv')
bh_df['Misinfo_Or_Books'] = 'misinfo'
bh_df['Title'] = 'ballotHarvesting'
bh_df = bh_df.drop(columns='i')
bh_df = bh_df.rename(columns={"name": "Location Name"})
```

```python
sts_df = pd.read_csv('misinfoNarratives/stopTheSteal/locationResults/results.csv')
```

```python
sts_df['Misinfo_Or_Books'] = 'misinfo'
sts_df['Title'] = 'stopTheSteal'
sts_df = sts_df.drop(columns='i')
sts_df = sts_df.rename(columns={"name": "Location Name"})
```

```python
df_dict = {}
def create_book_df(book: str):
    df = pd.read_csv(f'nytBestsellers2020/{book}/locations/results.csv')
    df['Title'] = book
    df_dict[book] = df
```

```python
create_book_df('accidentallyWesAnderson')
create_book_df('aRepublicUnderAssault')
create_book_df('battleOfBrothers')
create_book_df('doYouFeelLikeIDo')
create_book_df('greenLights')
create_book_df('howToBeAnAntiracist')
create_book_df('isThisAnything')
create_book_df('killingCrazyHorse')
create_book_df('oneVoteAway')
```

```python
book_df = pd.DataFrame(columns=['i', 'lat', 'lon', 'name', 'Title'])
```

```python
for key in df_dict:
    book_df = pd.concat([book_df, df_dict[key]], axis=0).reset_index(drop = True)
```

```python
book_df
```

```python
book_df['Misinfo_Or_Books'] = 'books'
book_df = book_df.drop(columns='i')
book_df = book_df.rename(columns={"name": "Location Name"})
```

```python
book_df
```

```python
book_df.to_csv('bookinfo_df.csv')
```

```python
joined_df = pd.concat([book_df, misinfo_df], axis=0).reset_index(drop = True)
```

```python
joined_df.to_csv('joined_df.csv')
```
