```python
from shapely.geometry import Point
from scipy.spatial import distance_matrix
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('../finalData/joined_df.csv')
```

```python
df['geometry'] = df.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
```

```python
gdf = gpd.GeoDataFrame(df, geometry='geometry')
```

```python
groups = gdf.groupby('Misinfo_Or_Books')
```

```python
misinfo = df[df['Misinfo_Or_Books']=='misinfo']
```

```python
most_common_locations = misinfo.groupby(['lat', 'lon']).size().reset_index(name='count').sort_values(by='count', ascending=False)

print(most_common_locations.head(10))
```

Most common misinformation locations
1. Dearing, Kansas
2. Winters, Texas
3. Avon Park, Florida
4. Gravesboro, California
5. New York, New York
6. Washington, DC
7. Level Corner, Pennsylvania
8. LA, California
9. Young, Arizona
10. Bells, North Carolina

```python
books = df[df['Misinfo_Or_Books']=='books']
```

```python
most_common_locations = books.groupby(['lat', 'lon']).size().reset_index(name='count').sort_values(by='count', ascending=False)

print(most_common_locations.head(10))
```
